# Make Train csv

```python
# f = open('./meta/train.txt')
f = open('./meta/test.txt')
files = f.readlines()
files = ['/host/data/dataset/food101/images/' + f.rstrip('\n') + '.jpg' for f in files]
```

```python
files
```

```python
from sys import exit

class_name_prev = 'hoge'
label = -1

for i, file in enumerate(files):
    class_name_now = file.split('/')[6]
    if class_name_now != class_name_prev:
        label += 1
        class_name_prev = class_name_now
    files[i] = '{},{}'.format(file, label)
    
        
#     class_name = file.split('/')[6]
#     exit()
```

```python
# f = open('train.csv', 'w')
f = open('test.csv', 'w')
for x in files:
    f.write(str(x) + "\n")
f.close()
```
